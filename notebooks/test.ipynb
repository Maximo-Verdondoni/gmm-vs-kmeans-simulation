```python
for escenario in [1,2,3,4]:
    #Generacion de datos
    X_test, y_real_test = generar_datos_escenario(escenario_id=escenario, N=500, random_state=42)

    print(f"Escenario: {escenario}")
    print(f"Dimensiones de X: {X_test.shape}")
    print(f"Dimensiones de y: {y_real_test.shape}")

    # K-Means
    kmeans_test = KMeans(n_clusters=3, random_state=42, n_init=10)
    y_pred_km = kmeans_test.fit_predict(X_test)
    centroides_km = kmeans_test.cluster_centers_

    # GMM
    gmm_test = GaussianMixture(n_components=3, covariance_type='full', random_state=42)
    y_pred_gmm = gmm_test.fit_predict(X_test)
    centroides_gmm = gmm_test.means_

    # Graficamos la comparativa
    plot_comparacion_modelos(X_test, y_pred_km, y_pred_gmm, centroides_km, centroides_gmm, 
                            titulo=f"Comparativa en Escenario {escenario}")
```
